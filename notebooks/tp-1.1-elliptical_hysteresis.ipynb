```python
import numpy as np
from matplotlib import pyplot as plt

"""
This script allows to plot the (elliptical) BH curve for a material with 
complex and permeability.
"""

mu0 = 4e-7*np.pi
h_amplitude = 100

```

```python
omega = np.linspace(0, np.pi/0.5, 1001)
amplitude_mu = 3000*mu0
phi_mus = [0, 10, 20]


h = h_amplitude*np.cos(omega)
b_linear = []
b_hysteresis = []
for phi_mu in phi_mus:
    b_hysteresis.append(amplitude_mu * h_amplitude * np.cos(omega-np.deg2rad(phi_mu)))


# BH curve transformed

for index, b_hyst in enumerate(b_hysteresis):
    plt.plot(h, b_hyst, "-", label=f"phi: {phi_mus[index]}°")
plt.grid()
plt.xlabel(r"H in A/m")
plt.ylabel(r"B / T")
plt.legend()
plt.show()
```
